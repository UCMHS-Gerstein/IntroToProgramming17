```python
class Rocket(object):
    """
    ATTRIBUTES
    x, y
    fuel
    speed
    guidance
    insulation
    
    METHODS
    move
    launch
    explode
    fly
    
    """
    
    def __init__(self, vel_y):
        self.x = 0
        self.y = 0
        self.vel_x = 1
        self.vel_y = vel_y
        self.fuel = 100
        self.g = -.098
    
    def move(self):
        """one unit of time passes each time we call this"""
        self.vel_y += self.g
        self.x += self.vel_x
        self.y += self.vel_y
        self.fuel -= .05
        if self.is_crashed():
            print("YOU CRASHED AND EVERYBODY IS DEAD")
#         self.status()

    def status(self):
        print(f"Rocket is at {self.x}, {self.y}")
        print(f"Velocity: {self.vel_x}, {self.vel_y}")
        print(f"Fuel Remaining: {round(self.fuel, 2)}%")
        print('---')
    
    def is_crashed(self):
        if self.y < 0:
            return True
        else:
            return False
```

```python
a = Rocket(100)
# b = Rocket()
```

```python
while not a.is_crashed():
    a.move()
    a.status()
```
